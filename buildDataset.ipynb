```python
def clean_tweet(text):
    text = text.replace("NEWLINE_TOKEN", "")
    text = text.lower()
    text = re.sub(r"http\S+",       "", text)  # remove URLs
    text = re.sub(r"@\w+",          "", text)  # remove mentions
    text = re.sub(r"#\S+",          "", text)  # remove hashtags like #stinking
    text = re.sub(r"[!?]+",         "", text)  # remove ! and ?
    text = re.sub(r"rt\s*:",        "", text)  # drop any RT:
    text = re.sub(r"&[^;\s]+?;",    "", text)  # drop HTML entities like &#8230;
    text = re.sub(r"\S+?:",         "", text)  # drop any word: anywhere
    text = re.sub(r'^[\s:]+',       "", text)  # strip leading colons/spaces
    text = re.sub(r'^"+',           "", text)  # drop leading quotes
    text = re.sub(r'"+$',           "", text)  # drop trailing quotes
    text = re.sub(r"\s+",           " ", text).strip()  # collapse whitespace & trim
    return text



df["clean_tweet"] = df["tweet"].apply(clean_tweet)
```
